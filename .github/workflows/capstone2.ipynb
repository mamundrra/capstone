**Drop the customer ID column**

```
df.drop('customerID', axis=1, inplace=True)
```

**Churn yes and not convert to 1 & 0**

```
df['Churn']=df['Churn'].map({'Yes':1,'No':0})
```

```
df.head()
```

**All category convert to encoder**

```
ceat_feat=['gender','Partner', 'Dependents', 'PhoneService', 'MultipleLines','InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']
```

```
pip install category_encoders
```

```
from category_encoders import OrdinalEncoder
```

```
ordinal_encoder=OrdinalEncoder(cols=ceat_feat, handle_missing='return_none')
```

```
data=ordinal_encoder.fit_transform(df)
```

```
data.head()
```

**Churn data corelation find**

```
def corr_for_target(data, target, title=None):
    plt.figure(figsize=(4,14))
    sns.set(font_scale=1)
    
    sns.heatmap(data.corr()[[target]].sort_values(target, ascending=False)[1:], annot=True, cmap="coolwarm")
```

```
corr_for_target(data,'Churn')
```

**Apply the model selection train test split**

```
from sklearn.model_selection import train_test_split
```

```
X=data.drop('Churn', axis=1)
```

```
y=data['Churn']
```

**Churn data need to data balanced**


```
from imblearn.over_sampling import SMOTE
```

```
oversample=SMOTE()
```

```
X, y=oversample.fit_resample(X, y)
```

```
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3, random_state=123)
```

**XGB Classifier algorithom apply**

```
from xgboost import XGBClassifier
```

```
model=XGBClassifier()
```

```
model.fit(X_train, y_train)
```

```
pred=model.predict(X_test)
```

```
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

**Model accuracy**

```
accuracy_score(pred, y_test)
```

```
accuracy_score(pred, y_test)
```

```
confusion_matrix(pred, y_test)
```

```
print(classification_report(pred, y_test))
```
